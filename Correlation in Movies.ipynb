```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
df = df.astype({'budget':'int'}) 
#Making columns of budget into all int instead of float
```

```python
df = df.dropna()
df['budget'].isnull().values.sum()
#Gonna drop all null values for budget because the data was incomplete, missing datas. 
```

```python
df['gross'].isnull().values.any()
```

```python
#change the data type of gross into int
df = df.astype({'gross':'int'}) 
```

```python
df['yearcorrect']=df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
#lets order this data by the highest gross to lowest 
df.sort_values(by = ['gross'],ascending=False ).head()
```

```python
#looking for correlation in data with budget and gross earning
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()




```

```python
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

```python
df.corr()
#Looking for some correlation only for numerical field. 
#We can see that with pearson method budget and gross have pretty high correlation
#Then we have votes with gross
```

```python
Matrix_correlation = df.corr()
sns.heatmap(Matrix_correlation, annot = True)
plt.title('Correlation Matrix for Numeric')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
#using seaborn heatmap for better visualization with colors
```

```python
#We can numerize the object/string into catagory then convert it into numeric value for our
#corresponding matrix as df.corr() only take numeric value 
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()


```

```python
#Lets see the new matrix we formed 
Matrix_correlation = df_numerized.corr()
sns.heatmap(Matrix_correlation, annot = True)
plt.title('Correlation Matrix for Numeric')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_numerized.corr()
```

```python
#Or we can see this in an easier way by using unstack for each variable/movie features and sort them
correlation_matrix = df_numerized.corr()
correlation_pairs = correlation_matrix.unstack()
correlation_pairs = correlation_pairs.sort_values()
high_correlation = correlation_pairs[(correlation_pairs) > 0.4  ]
high_correlation
```
